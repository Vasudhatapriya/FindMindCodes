```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
# import datetime
import time
import os


# Specify the path where chromedriver is installed
chrome_path = r"C:\Users\Vashudha Tapriya\Desktop\chromedriver_win32\chromedriver.exe"

#website that has to be scraped
driver = webdriver.Chrome(chrome_path)
driver.get("https://bombayhighcourt.nic.in/")

#getting on the correct page
link=driver.find_element(By.XPATH,'/html/body/div[2]/table/tbody/tr[2]/th/table/tbody/tr/td/div/ul[4]/li/a')
link.click()
time.sleep(2)
link1=driver.find_element(By.XPATH,'/html/body/div[2]/table/tbody/tr[2]/th/table/tbody/tr/td/div/ul[4]/li/ul/li[2]/a')
link1.click()
time.sleep(2)
link2=driver.find_element(By.XPATH,'/html/body/div[2]/table/tbody/tr[2]/th/table/tbody/tr/td/div/ul[4]/li/ul/li[2]/ul/li[3]/a')
link2.click()
time.sleep(2)
prewindow = driver.window_handles[0]

count1=10
count2=0


def get_data():
    count1=10
    count2=0
    prewindow = driver.window_handles[0]
    CORUM,PARTY,JUDGEMENT_DATE_BENCH,CASENO=[],[],[],[]
    #selecting Side
    for count in range (1,8):
            
        #clearing boxes
        datef=driver.find_element(By.XPATH,'/html/body/form/table[4]/tbody/tr/td[2]/input')
        datet=driver.find_element(By.XPATH,'/html/body/form/table[4]/tbody/tr/td[5]/input')
        datef.clear()
        datet.clear()
        
        driver.switch_to.window(prewindow)
        opt = driver.find_element(By.XPATH,'/html/body/form/table[2]/tbody/tr/td[2]/select/option['+str(count)+']')
        opt.click()
        time.sleep(2)

        #selecting the ACT
        act=driver.find_element(By.XPATH,'/html/body/form/table[3]/tbody/tr/td[3]/select')
        drp=Select(act)
        drp.select_by_visible_text('All Act')
        time.sleep(2)

        #from date automation selection
        sdate = driver.find_element(By.XPATH,'/html/body/form/table[4]/tbody/tr/td[3]/a/img')
        prewindow = driver.window_handles[0]
        sdate.click()
        newwindow = driver.window_handles[1]
        driver.switch_to.window(newwindow)

        #from the month of start date
        month=driver.find_element(By.XPATH,'/html/body/form/table/tbody/tr[1]/td/table/tbody/tr/td[1]/select')
        m=Select(month)
        m.select_by_visible_text("January")


        #from the year of start date
        for i in range(10-count2):    
            year=driver.find_element(By.XPATH,'/html/body/form/table/tbody/tr[1]/td/table/tbody/tr/td[2]/a[1]')
            year.click()

        #from the day of start date 
        day=driver.find_element(By.XPATH,'/html/body/form/table/tbody/tr[4]/td[5]/font/a')
        time.sleep(2)
        day.click()

        driver.switch_to.window(prewindow)


        #to date automation selection
        edate = driver.find_element(By.XPATH,'/html/body/form/table[4]/tbody/tr/td[6]/a/img')
        edate.click()

        newwindow = driver.window_handles[1]
        driver.switch_to.window(newwindow)


        #to the month of start date
        month=driver.find_element(By.XPATH,'/html/body/form/table/tbody/tr[1]/td/table/tbody/tr/td[1]/select')
        m=Select(month)
        m.select_by_visible_text("December")

        #from the year of start date
        for i in range(10-count2):    
            year=driver.find_element(By.XPATH,'/html/body/form/table/tbody/tr[1]/td/table/tbody/tr/td[2]/a[1]/b/font')
            year.click()

        #to date 
        day=driver.find_element(By.XPATH,'/html/body/form/table/tbody/tr[8]/td[5]/font/a')
        day.click()
        count2=count2+1
        
        #captcha
        captcha=input("Enter the CAPTCHA: ")
        time.sleep(10)

        box=driver.find_element(By.XPATH,'/html/body/form/table[6]/tbody/tr/td[2]/input')
        box.send_keys(captcha)
        
        #GO
        go=driver.find_element(By.XPATH,'/html/body/form/table[7]/tbody/tr/td[1]/input')
        go.click()

        ar = driver.find_element_by_xpath('/html/body/form/table[2]/tbody').get_attribute(('innerHTML'))
        table = BeautifulSoup(ar, "lxml")

        for links in table.find_all('a'):
            link = links["href"] 
            CASENO.append("https://bombayhighcourt.nic.in/" + link)
        row_count = len(CASENO)

        for i in range(1,row_count+1):

            y = (driver.find_element_by_xpath('/html/body/form/table[2]/tbody/tr[' + str(i) + ']/td[1]')).text
            CORAM.append(y)

            z = (driver.find_element_by_xpath('/html/body/form/table[2]/tbody/tr[' + str(i) + ']/td[2]')).text
            PARTY.append(z)

            w = (driver.find_element_by_xpath('/html/body/form/table[2]/tbody/tr[' + str(i) + ']/td[3]')).text
            JUDGEMENT_DATE_BENCH.append(w)

            q = (driver.find_element_by_xpath('/html/body/form/table[2]/tbody/tr[' + str(i) + ']/td[4]')).text
            CASENO.append(q)

        dictionary = [{"SrNo" : val[0],
                       "CaseType" : val[1], 
                       "OrderDate" : val[2], 
                        "Ordernolink" : val[3]}
                      for val in zip(CORAM,PARTY,JUDGEMENT_DATE_BENCH,CASENO)]
        return dictionary 


dictionary=get_data()
print(dictionary)
    
```
