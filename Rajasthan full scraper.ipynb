```python
# from PIL import Image
# import cv2
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
# import datetime
import time
import os


# Specify the path where chromedriver is installed
chrome_path = r"C:\Users\Vashudha Tapriya\Desktop\chromedriver_win32\chromedriver.exe"

#website that has to be scraped
driver = webdriver.Chrome(chrome_path)
driver.get("https://services.ecourts.gov.in/ecourtindiaHC/cases/s_orderdate.php?state_cd=9&dist_cd=1&court_code=1&stateNm=Rajasthan")



#selecting the year of start date
def get_data():
    pdf_link,SrNo,CaseType,OrderDate=[],[],[],[]
    for i in range(2010,2019):
        
        #starting date automation selection
        sdate = driver.find_element(By.XPATH,'/html/body/form/div[2]/div[1]/span[3]/img')
        sdate.click()
        time.sleep(2)

        year=driver.find_element(By.XPATH,'/html/body/div[4]/div/div[2]/div/div/select[2]')
        y=Select(year)
        y.select_by_visible_text(str(i))
        # year.click()   

        #selecting the month of start date
        month=driver.find_element(By.XPATH,'/html/body/div[4]/div/div[2]/div/div/select[1]')
        m=Select(month)
        m.select_by_visible_text("January")

        #selecting the day of start date
        date = driver.find_element(By.XPATH,'/html/body/div[4]/div/div[2]/div/table/tbody/tr[1]/td[6]/a')
        date.click()


        #ending date automation selection(after 1 year)
        edate = driver.find_element(By.XPATH,'/html/body/form/div[2]/div[2]/span[3]/img')
        edate.click()
        time.sleep(2)

        #selecting the month of end date
        monthe=driver.find_element(By.XPATH,'/html/body/div[4]/div/div[2]/div/div/select[1]')
        me=Select(monthe)
        me.select_by_visible_text("December")

        #selecting the year of end date
        yeare=driver.find_element(By.XPATH,'/html/body/div[4]/div/div[2]/div/div/select[2]')
        ye=Select(yeare)
        ye.select_by_visible_text(str(i))
        # year.click()

        #selecting the day of end date
        date = driver.find_element(By.XPATH,'/html/body/div[4]/div/div[2]/div/table/tbody/tr[5]/td[6]/a')
        date.click()
        time.sleep(2)
        driver.maximize_window()
        time.sleep(5)

        captcha=input("Enter the CAPTCHA: ")
        time.sleep(5)

        box=driver.find_element(By.XPATH,'/html/body/form/div[2]/div[3]/span/div[2]/span[3]/input')
        box.send_keys(captcha)
        
        
        #selecting the go button
        go=driver.find_element(By.XPATH,'/html/body/form/div[2]/div[4]/span[3]/input[1]')
        go.click()
        print("going to sleep")
        time.sleep(40)   
        print("came back")
        # #downloading the captcha image 
        # driver.find_element_by_id("captcha_image")
        # driver.get_screenshot_as_file('C:/Users/Vashudha Tapriya/Desktop/screenshot/google.png')
        # time.sleep(2)

        # #to calculate coordinates
        # # im = cv2.imread("C:/Users/Vashudha Tapriya/Desktop/screenshot/google.png")
        # # r = cv2.selectROI(im)
        # # print(int(r[1]), int(r[1]+r[3]), int(r[0]), int(r[0]+r[2]))

        # # Read image
        # img=Image.open("C:/Users/Vashudha Tapriya/Desktop/screenshot/google.png")
        # img2=img.crop(((525,445,851,639)))
        # img2.save("C:/Users/Vashudha Tapriya/Desktop/screenshot/img2.png")

        ar = driver.find_element_by_xpath('//*[@id="showList3"]').get_attribute(('innerHTML'))
        table = BeautifulSoup(ar, "lxml")
        
        for links in table.find_all('a'):
            link = links["href"]
            pdf_link.append("https://services.ecourts.gov.in/" + link)
        row_count = len(pdf_link)

        for i in range(1,row_count+1):

            y = (driver.find_element_by_xpath('//*[@id="showList1"]/tr[' + str(i) + ']/td[1]')).text
            SrNo.append(y)
            
            z = (driver.find_element_by_xpath('//*[@id="showList1"]/tr[' + str(i) + ']/td[2]')).text
            CaseType.append(z)

            w = (driver.find_element_by_xpath('//*[@id="showList1"]/tr[' + str(i) + ']/td[3]')).text
            OrderDate.append(w)

            q = (driver.find_element_by_xpath('//*[@id="showList1"]/tr[' + str(i) + ']/td[4]')).text
            pdf_link.append(q)
        
        dictionary = [{"SrNo" : val[0],
                       "CaseType" : val[1], 
                       "OrderDate" : val[2], 
                        "Ordernolink" : val[3]}
                      for val in zip(SrNo,CaseType,OrderDate, pdf_link)]
        return dictionary 
    
      
dictionary=get_data()
print(dictionary)
```
