```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
# import datetime
import time
import os


# Specify the path where chromedriver is installed
chrome_path = r"C:\Users\Vashudha Tapriya\Desktop\chromedriver_win32\chromedriver.exe"

#website that has to be scraped
driver = webdriver.Chrome(chrome_path)
driver.get("http://cms.nic.in/ncdrcusersWeb/search.do?method=loadSearchPubLndc")

#starting date automation selection
sdate = driver.find_element(By.XPATH,'/html/body/form/div/div[2]/table/tbody/tr[2]/td[2]/input')
sdate.click()
time.sleep(2)

#selecting the year of start date
year=driver.find_element(By.XPATH,'/html/body/div/div/div[2]/div/div/select[2]')
y=Select(year)
y.select_by_value("1/2019")
# year.click()

#selecting the month of start date
month=driver.find_element(By.XPATH,'/html/body/div/div/div[2]/div/div/select[1]')
m=Select(month)
m.select_by_visible_text("January")

#selecting the day of start date
date = driver.find_element(By.XPATH,'/html/body/div/div/div[2]/div/table/tbody/tr[1]/td[3]/a')
date.click()


#ending date automation selection(after 1 year)
edate = driver.find_element(By.XPATH,'/html/body/form/div/div[2]/table/tbody/tr[2]/td[3]/input')
edate.click()
time.sleep(2)

#selecting the month of end date
monthe=driver.find_element(By.XPATH,'/html/body/div/div/div[2]/div/div/select[1]')
me=Select(monthe)
me.select_by_visible_text("January")

#selecting the year of end date
yeare=driver.find_element(By.XPATH,'/html/body/div/div/div[2]/div/div/select[2]')
ye=Select(yeare)
ye.select_by_value("1/2020")
# year.click()


#selecting the day of end date
date = driver.find_element(By.XPATH,'/html/body/div/div/div[2]/div/table/tbody/tr[1]/td[4]/a')
date.click()

#selecting the option for between dates
searchby=driver.find_element(By.XPATH,'/html/body/form/div/div[2]/table/tbody/tr[2]/td[4]/select')
drp=Select(searchby)
drp.select_by_visible_text('Between dates')

#selecting the select commission
selcommision1=driver.find_element(By.XPATH,'/html/body/form/div/div[2]/table/tbody/tr[2]/td[5]/select')
drp=Select(selcommision1)
drp.select_by_visible_text('NCDRC')


#selecting the SearchIn 
selcommision2=driver.find_element(By.XPATH,'/html/body/form/div/div[2]/table/tbody/tr[2]/td[6]/select')
drp1=Select(selcommision2)
drp1.select_by_visible_text('NCDRC')

```
